**AUTOMATED DATA PROCESSING PIPELINE**
============================================

**Goals:** Automate the transformation of uncleaned data into a uniform format using Python, thereby reducing the manual data entry workload.

**Tool used:** Python (Pandas library)

**How does it work:**
This Python script processes data from an Excel file named VNM.xlsx using the pandas library.

It reads the data, then iterates through each row. If the 'Interview result' column is not empty, it creates two new rows with a specific format. The first new row contains data from the original row (such as Duns Number, names, and interview details) and is tagged for a "DELETEADD" action. The second new row contains hardcoded information about the data source.

Finally, the script creates a new Excel file named processed_data.xlsx with two sheets: one for the original data and one for the newly created rows.


```python
import pandas as pd 

df = pd.read_excel('VNM.xlsx') 

new_rows = []

print(df.head())

DUNS = 'Duns_No'
VALUE = 'Interview result'
DATE = 'Interview date'
FS = 'Family surname'
VFS = 'Vie family surname'
PO = 'Position [String]'
POS = 'PostTitle [String]'
MODE = 'Mode of interview'
PHONE = 'Phone'

for index, row in df.iterrows():
    if pd.notna(row[VALUE]) and str(row[VALUE]).strip() != '':
        new_row1 = {
            'ID [String]':' ',
            'DUNS [String]': row[DUNS],
            'CrifNumber [String]':'',
            'DateOfInterview [DateTime]': row[DATE],
            'NameOfInterviewed [String]':' ',
            'NameOfInterviewedLocal [String]':' ',
            'SurnameOfInterviewed [String]': row[FS],
            'SurnameOfInterviewedLocal [String]': row[VFS],
            'FlagSurnameFirst [Boolean]':' ',
            'Qualifier [String]':' ',
            'PreTitle [String]':' ',
            'Position [String]': row[PO],
            'PostTitle [String]': row[POS],
            'InterviewResults [String]': row[VALUE],
            'ModeOfInterview [String]': row[MODE],
            'AdditionalComments [String]':' ',
            'AdditionalCommentsLocal [String]':' ',
            'FlagUnsuccessfulAttemptsMade [Boolean]':' ',
            'Action [String]':'DELETEADD'}

        new_row2 = {
             'ID [String]':' ',
            'DUNS [String]': row[DUNS],
            'CrifNumber [String]':'',
            'DateOfInterview [DateTime]':'18/10/2024',
            'NameOfInterviewed [String]':' ',
            'NameOfInterviewedLocal [String]':' ',
            'SurnameOfInterviewed [String]':' ',
            'SurnameOfInterviewedLocal [String]':' ',
            'FlagSurnameFirst [Boolean]':' ',
            'Qualifier [String]':' ',
            'PreTitle [String]':' ',
            'Position [String]':' ',
            'PostTitle [String]':' ',
            'InterviewResults [String]':' ',
            'ModeOfInterview [String]':' ',
            'AdditionalComments [String]':'On 18/10/2024, the information in this report is obtained from Local registry office',
            'AdditionalCommentsLocal [String]':'Vào ngày 18 tháng 10 năm 2024, thông tin trong báo cáo này được lấy từ văn phòng đăng ký địa phương',
            'FlagUnsuccessfulAttemptsMade [Boolean]':' ',
            'Action [String]':' '}
        new_rows.append(new_row1)
        new_rows.append(new_row2)

new_rows_df = pd.DataFrame(new_rows)

with pd.ExcelWriter('processed_data.xlsx') as writer:
    df.to_excel(writer, sheet_name='Original Data', index=False)
    new_rows_df.to_excel(writer, sheet_name='New Rows', index=False)
```
